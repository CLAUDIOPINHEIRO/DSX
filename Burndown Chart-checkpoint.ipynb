```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Dias do Sprint
dias_nomes = ['seg', 'ter', 'qua', 'qui', 'sex', 'seg', 'ter', 'qua', 'qui', 'sex']
dias_num = range(1,11)
```

```python
## Historias e Pontos
historias = [
    {'id': 1, 'pontos': 5, 'desc': "Deploy de Prod SFP"},
    {'id': 2, 'pontos': 13, 'desc': "Job de Promoção SFP"},
    {'id': 1, 'pontos': 8, 'desc': "Backup Automatico Homolog"},
    {'id': 1, 'pontos': 3, 'desc': "Build do Batch XPTO"},
    {'id': 1, 'pontos': 5, 'desc': "Testes Selenium ARV Homolog"}
]
pontos = 0
for historia in historias:
    pontos+= historia['pontos']
linha_de_pontos = [pontos]    
```

```python
## Historias concluidas por dia
pontos_queimados = [0, 8, 13, 5, 3, 5]

## Atualiza os pontos queimados
for ponto in pontos_queimados:
    linha_de_pontos.append(linha_de_pontos[-1] - ponto)
```

```python
## Preenche os dias restantes
linha_de_pontos = linha_de_pontos + [linha_de_pontos[-1]] * (len(dias_nomes) - len(linha_de_pontos))
```

```python
plt.figure(figsize=(16, 9))
plt.plot(dias_num, linha_de_pontos)
plt.xticks(dias_num, dias_nomes)
plt.ylim(0, pontos + 2)
plt.xlabel("Dias do Sprint")
plt.ylabel("Pontos Queimados")
plt.title("Burndown Chart - Sprint #4")
plt.show()
```
